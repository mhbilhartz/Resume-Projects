```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('../../../Downloads/Lab 03 Repeatability Bilhartz-Lopez(Sheet1) (1).csv')
print(df)
values = df.values.flatten()
values = values[~pd.isnull(values)]  # removes NaN
print(values)
```

```python
# Plot histogram

plt.figure(figsize=(12,6))
plt.hist(values, bins=15, edgecolor="black", alpha=0.7)
plt.title("Histogram of (mV) Measurements on Kafadar - Lab 03 Repeatibility",  fontsize= 16)
plt.xlabel("Milli Volts", fontsize= 12)
plt.xticks(rotation=45, fontsize=12)
plt.ylabel("Number of measurements",  fontsize= 12)

plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python

# Flatten + clean NaN
#values = df_2.values.flatten()
#values = values[~pd.isnull(values)]

# Key statistics
mean_val = np.mean(values)
median_val = np.median(values)
mode_val = stats.mode(values, keepdims=True)[0][0]   # scipy returns array
q1 = np.percentile(values, 25)
q3 = np.percentile(values, 75)
iqr = q3 - q1

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
print("Q1 (Lower Quartile):", q1)
print("Q3 (Upper Quartile):", q3)
print("IQR:", iqr)

# Assuming your data is stored in values (a NumPy array)
N = len(values)                       # number of samples
std_val = values.std(ddof=1)          # sample standard deviation
sem = std_val / np.sqrt(N)            # standard error of the mean

print("Number of samples (N):", N)
print("Standard Deviation:", std_val)
print("Standard Error of the Mean:", sem)
```
